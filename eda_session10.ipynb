**Data transformations**

-generally data has 3 types of data transformations
     -normal distribution
     -positive skewed
     -negative skewed
-all the math conceptsndeveloped by assumption as Data follows normal distribution

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
exp_data=np.random.exponential(size=10000)
exp_data
```

```python
plt.hist(exp_data,bins=20,label='Exponential')
```

norm_data=np.random.normal(size=10000)
norm_data

```python
plt.hist(norm_data,bins=20,label='Exponential')
```

**LOG TRANSFORMATION**

-means performing logarithm operations on original data
-it is an approach to convrt into normal distribution
-log means natural logarithm 

```python
x=2
np.log(x)
```

```python
exp_norm=np.log(exp_data)
```

```python
exp_norm
```

```python
plt.hist(exp_norm,bins=20,label='logdata')
```

```python
plt.subplot(1,2,1)
plt.hist(exp_data,bins=20,label='Exponential')
plt.subplot(1,2,2)
plt.hist(exp_norm,bins=20,label='logdata')
```

**reciprocal transformatio**

```python
x=2
1/x
```

```python
exp_data=np.random.exponential(size=10000)
reci=1/exp_data
```

```python
reci
```

```python
plt.subplot(1,2,1)
plt.hist(exp_data,bins=20,label='Exponential')
plt.subplot(1,2,2)
plt.hist(reci,bins=20,label='reciprocal')
plt.show()
```

**SQUARE ROOT**

```python
x=4
y=np.sqrt(x)
```

```python
y
```

```python
sqr=np.sqrt(exp_data)
sqr
```

```python
plt.subplot(1,2,1)
plt.hist(exp_data,bins=20,label='Exponential')
plt.subplot(1,2,2)
plt.hist(sqr,bins=20,label='square root')
plt.show()
```
