```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
visa_df=pd.read_csv('visa.csv')
visa_df
```

```python
l1=['prevailing_wage','yr_of_estab','no_of_employees']
prev=[]
est=[]
emp=[]
for i in l1:
    wage_count=round(len(visa_df[i]),2)
    wage_min=round(visa_df[i].min(),2)
    wage_max=round(visa_df[i].max(),2)
    wage_mean=round(visa_df[i].mean(),2)
    wage_median=round(visa_df[i].median(),2)
    wage_std=round(visa_df[i].std(),2)
    list_values=[wage_count,wage_min,wage_max,
    wage_mean,wage_median,wage_std]
    index_val=['count','min','max','mean','median','std']
    l2=[wage_count,wage_min,wage_max,wage_mean,wage_median,wage_std]
    for j in l2:
        
      if i=='prevailing_wage':
        prev.append(j)
      elif i=='yr_of_estab':
        est.append(j)
      elif i=='no_of_employees':
        emp.append(j)
pd.DataFrame(zip(prev,est,emp),columns=l1,index=index_val)
        
        
```
